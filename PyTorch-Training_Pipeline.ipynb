### Converting to Tensors

```python
x_train_ten=torch.from_numpy(x_train)
x_test_ten=torch.from_numpy(x_test)
y_train_ten=torch.from_numpy(y_train)
y_test_ten=torch.from_numpy(y_test)
```

### Defining the model 

```python
class MySimpleNN():
    def __init__(self,X):         # defining the archietecture of the NN
        self.weights = torch.rand(X.shape[1],1,dtype=torch.float64,requires_grad=True)
        self.bias = torch.zeros(1,dtype=torch.float64,requires_grad=True)
        
    def forward_pass(self,X):       # Forward pass
        z=torch.matmul(X,self.weights) + self.bias
        y_pred = torch.sigmoid(z)

        return y_pred

    def loss_func(self,y_pred,Y):  # Loss calcuation per epoch
        ep=1e-7
        y_pred=torch.clamp(y_pred,ep,1-ep)
        loss= -(Y*torch.log(y_pred) + (1-Y)*torch.log(1-y_pred)).mean()
    
        return loss

    def parameter_update(self):    # Parameter Updation
        loss.backward()
```

```python
### important parameters

learning_rate=0.1
epochs=25
```

```python
## creating the model 

model= MySimpleNN(x_train_ten)
```

```python
model.weights
```

```python
model.bias
```

```python
### training the model 

for epoch in range(epochs):

    # forward pass 
    y_pred=model.forward_pass(x_train_ten)
  

    # loss 
    loss=model.loss_func(y_pred,y_train_ten)

    # backward pass
    model.parameter_update()

    # parametr updation
    with torch.no_grad():
        model.weights -= learning_rate * (model.weights.grad)
        model.bias-= learning_rate * (model.bias.grad)

    # zero gradients 
    model.weights.grad.zero_()
    model.bias.grad.zero_()

    print(f" for EPOCH {epoch+1} => LOSS = {loss}")
```

```python
with torch.no_grad():
    y_pred=model.forward_pass(x_test_ten)
    y_pred=(y_pred>0.5).float()
    print(y_pred,y_test_ten)
    
    accuracy=(y_pred == y_test_ten).float().mean()
    
    print(f"Accuracy : {accuracy}")  
```
